```python
import os
import numpy as np
import keras
import matplotlib.pyplot
from tensorflow import data

train , val = keras.utils.image_dataset_from_directory("petimages",
                                                       validation_split=0.2,
                                                       subset='both',
                                                       seed=902,
                                                       image_size=(180, 180),
                                                       batch_size=128)


data_aug_layers = [keras.layers.RandomFlip('Horizontal'),
                   keras.layers.RandomRotation(0.2)]

def data_aug(images):
    for layer in data_aug_layers:
        images = layer(images)
        return images
    
```
